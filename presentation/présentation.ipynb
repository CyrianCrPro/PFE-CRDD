```python
import os
import matplotlib.pyplot as plt
import cv2

# Define path to training dataset
train_dir = "./gtsrb-german-traffic-sign/versions/1/Train"  # Update this with the actual path

# Get class names (folder names)
class_folders = sorted(os.listdir(train_dir))

# Dictionary to store first image of each class
first_images = {}

for class_name in class_folders:
    class_path = os.path.join(train_dir, class_name)
    
    # Ensure it's a directory
    if os.path.isdir(class_path):
        # List all image files in the directory
        images = [f for f in os.listdir(class_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        if images:
            # Get the first image
            first_image_path = os.path.join(class_path, images[0])
            first_images[class_name] = first_image_path

# Display the first image from each class
plt.figure(figsize=(20, 10))
for i, (class_name, img_path) in enumerate(first_images.items()):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    plt.subplot(2, len(first_images)//2 + 1, i + 1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Class {class_name}")

plt.tight_layout()
plt.show()
```
